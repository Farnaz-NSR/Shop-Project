```python
import requests
from bs4 import BeautifulSoup

def crawl_zoomit():
    laptops = []
    display= []
    ram = []
    cpu = []
    brand = []
    price = []
    gpu = []

    url = "https://www.zoomit.ir/product/list/laptop/"

    # Headers to simulate a browser visit
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }

    try:
        # Send a GET request
        response = requests.get(url, headers=headers)
        response.raise_for_status()

        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all elements with the specified class
        elements = soup.find_all('a', class_="sc-f3acbc5d-0 dvpmzx")
        el2 = soup.find_all("span", class_="sc-68673876-0 hzaOrg sc-9f4b479e-1 gjiMjU") 
        el_price = soup.find_all("span", class_="sc-68673876-0 eouUcl fa")
        el_gpu = soup.find_all("span", class_="sc-68673876-0 eisdEk sc-9f4b479e-1 gjiMjU")

        # Extract and display the text and links
        for element in elements:
            text = element.get_text(strip=True)
            for item in brands:
                if item in text:
                    brand.append(item)
            laptops.append(text)
        
        counter = 0
        for el in el2:
            text = el.get_text(strip=True)
            if counter==0:
                display.append(text.split(" ")[0])
                counter = counter+1
            elif counter==1:
                cpu.append(text)
                counter = counter+1
            elif counter==2:
                ram.append(text.split(" ")[0])
                counter = counter+1
            elif counter==3:
                counter = 0

        for el in el_price:
            text = el.get_text(strip=True)
            price.append(text)



        counter = 1
        while counter <= len(el_gpu):
                text = el_gpu[counter].get_text(strip=True)
                gpu.append(text)
                counter = counter + 4

    except requests.exceptions.RequestException as e:
        print(f"Error occurred: {e}")
    
    return laptops, display, brand, ram, cpu, price, gpu
```

```python
laptops, display, brand, ram, cpu, price, gpu = crawl_zoomit()
```

```python
laps = {'brand': brand, 'cpu': cpu, 'gpu': gpu, 'ram': ram, 'display': display, 'price': price}
#create DataFrame
df = pd.DataFrame(laps)
df.to_csv('data/products.csv')
```
